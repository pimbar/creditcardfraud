```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
var = [c for c in df.columns if not c in ['Class']]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(df[var])
array_scaled = scaler.transform(df[var])
```

```python
df_scaled=pd.DataFrame(data=array_scaled, columns=var)
```

```python
df_scaled = pd.concat((df_scaled, df['Class']), axis=1)
```

```python
df_scaled
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df_scaled, test_size=0.3, stratify=df['Class'], random_state=999)
```

```python
# Downsample majority class
```

```python
from sklearn.utils import resample
df_majority = train[train.Class==0]
df_minority = train[train.Class==1]
 
# Downsample majority class
df_majority_downsampled = resample(df_majority, 
                                 replace=False,    # sample without replacement
                                 n_samples=100000,     # to match minority class
                                 random_state=999) # reproducible results
 
# Combine minority class with downsampled majority class
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
 
# Display new class counts
df_downsampled.Class.value_counts()
```

```python
df_majority = df_downsampled[df_downsampled.Class==0]
df_minority = df_downsampled[df_downsampled.Class==1]
 
# Upsample minority class
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=100000,    # to match majority class
                                 random_state=999) # reproducible results
 
# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
 
# Display new class counts
df_upsampled.Class.value_counts()
```

```python
X_train = df_upsampled[var]
y_train = df_upsampled['Class']
X_test = test[var]
y_test = test['Class']
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=500, random_state=999, max_depth=8, criterion='entropy',
                               n_jobs=-1, min_samples_leaf=10)
```

```python
model.fit(X_train, y_train)
```

```python
predict = model.predict_proba(X_test)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, model.predict_proba(X_test)[:,1])
print (auc(false_positive_rate, true_positive_rate))
print (roc_auc_score(y_test, model.predict(X_test)))
```

```python
from sklearn.metrics import recall_score
print(recall_score(y_test,model.predict(X_test)))
```
